```python
import pandas as pd
import seaborn as sb

titanic = pd.read_csv('titanic.csv')
```

```python
print (titanic['deck'].head)
```

```python
ponti = titanic['deck'].dropna().unique() #Tolgo i valori null e scopro quali ponti ci sono
print(f"Ponti presenti: {sorted(ponti)}")
print(f"Numero totale di ponti diversi: {len(ponti)}")
```

```python
#Visualizzare a grafico il numero di passeggeri per ogni classe d'imbarco.
#Conto quanti passeggeri per classe
conteggio_classi = titanic['pclass'].value_counts().sort_index()
print(conteggio_classi)
```

```python
conteggio_classi.plot(kind="bar") 
```

```python
#Conto i sopravvissuti
conteggio_survived = titanic['survived'].value_counts().sort_index()
conteggio_survived.plot(kind='bar', color='lightgreen') #Aggiungo un dettaglio grafico come il colore
```

```python
#Importo la libreria matplot e mostro la distribuzione delle tariffe
import matplotlib.pyplot as plt

sb.histplot(titanic['fare'], bins=30, kde=True, color='skyblue')
plt.xlabel('Tariffa (Fare)')
plt.ylabel('Numero di passeggeri')
plt.title('Distribuzione delle tariffe pagate dai passeggeri (con KDE)')
plt.show()
```

```python
#Ora invece di un istogramma creo un boxplot
sb.boxplot(x="pclass", y="age", data=titanic)

plt.xlabel("Classe di imbarco (Pclass)")
plt.ylabel("Età dei passeggeri")
plt.title("Distribuzione delle età rispetto alla classe di imbarco")
plt.show()
```

```python
sb.boxplot(x="survived", y="fare", data=titanic)

plt.xlabel("Sopravvissuto (0 = No, 1 = Sì)")
plt.ylabel("Tariffa pagata (Fare)")
plt.title("Distribuzione delle tariffe pagate rispetto alla sopravvivenza")
plt.show()
```

```python
#Si nota una mediana più alta per i sopravvissuti
#Se la mediana della tariffa pagata da chi è sopravvissuto è più alta, può indicare che i passeggeri che pagavano tariffe maggiori (magari di classi più alte) avevano maggiori probabilità di sopravvivere.
#Si può vedere che chi è sopravvissuto ha una gamma più ampia di tariffe, inclusi i valori molto alti (outlier).
#Conclusione le differenze nella distribuzione può mostrare che la classe sociale (indirettamente rappresentata da fare) ha influenzato la sopravvivenza.
```
